```python
import pandas as pd
import numpy as np
```

```python
microbiome=pd.read_table("../../haddad_data/raw/otu_table_haddad_fecal__expected_duration_6weeks__.txt", sep='\t', skiprows=1, index_col="#OTU ID")

metabolome=pd.read_table("../../haddad_data/raw/haddad_master_featuretable_unnorm.txt", sep='\t', header=0, index_col="#OTU ID")

metadata=pd.read_table("../../haddad_data/raw/haddad_master_metadata_fecal_w_alphametrices__expected_duration_6_weeks__.txt", sep='\t', header=0,index_col="#SampleID")
```

```python
#clean micro
microbiome=microbiome.T
microbiome=microbiome.reset_index().rename(columns={"index":"#SampleID"})
microbiome=microbiome.set_index("#SampleID")
```

```python
#clean metab
metabolome=metabolome.T
metabolome=metabolome.reset_index().rename(columns={"index":"#SampleID"})
metabolome=metabolome.set_index("#SampleID")
```

```python
def match_samples(mic, met, md):
    mic_samp=set(mic.index)
    met_samp=set(met.index)
    md_samp=set(md.index)
    matched=sorted(list(mic_samp & met_samp & md_samp))
    mic=mic.ix[matched]
    met=met.ix[matched]
    md=md.ix[matched]
    return(mic, met, md)
    
    import unittest
    m1 = pd.DataFrame([[0, 0, 1],
                  [2, 2, 4],
                  [5, 5, 3]],
                  index=['s1', 's2', 's3'],
                  columns=['o1', 'o2', 'o3'])
    _mapping = pd.DataFrame([['a', 'control'],
                        ['b', 'control'],
                        ['c', 'diseased'],
                        ['d', 'diseased']],
                        index=['s1', 's2', 's3', 's4'],
                        columns=['Barcode', 'Treatment'])
    m2=pd.DataFrame([[2, 4, 4],
                 [5, 3, 3],
                 [0, 0, 1]],
                index=['s2', 's3', 's4'],
                columns=['c1', 'c2', 'c3'])
    assert(match_samples(m1, m2, _mapping)[0].index==['s2','s3'])
```

```python
microbiome, metabolome, metadata=match_samples(microbiome, metabolome, metadata)
```

```python
print(microbiome.shape)
print(metabolome.shape)
print(metadata.shape)
```

#### filter zero features


```python
frem=[]
for otu in microbiome.columns:
    ll=np.array(microbiome[otu])
    if np.count_nonzero(ll) == 0:
        frem.append(otu)
microbiome=microbiome.drop(frem, axis=1)
```

```python
frem=[]
for mol in metabolome.columns:
    ll=np.array(metabolome[mol])
    if np.count_nonzero(ll) == 0:
        frem.append(mol)
metabolome=metabolome.drop(frem, axis=1)
```
