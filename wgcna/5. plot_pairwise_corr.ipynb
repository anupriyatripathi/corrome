```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#filter based on p-values
```

```python
corrmat=pd.read_table("./sotu_metab_corrmat_lc_pr.tsv", sep='\t', index_col=0, header=0)
```

```python
cm=sns.clustermap(corrmat.T, metric='correlation')
hm = cm.ax_heatmap.get_position()
plt.setp(cm.ax_heatmap.yaxis.get_majorticklabels(), fontsize=25, rotation=0)
plt.setp(cm.ax_heatmap.xaxis.get_majorticklabels(), fontsize=20, rotation=90)
cm.ax_heatmap.set_position([hm.x0, hm.y0, hm.width*2, hm.height*5])
col = cm.ax_col_dendrogram.get_position()
cm.ax_col_dendrogram.set_position([col.x0, col.y0, col.width*2, col.height*5])
plt.show()
```
