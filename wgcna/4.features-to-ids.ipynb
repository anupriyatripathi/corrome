## fuzzy conversion of MS1 features to MS2-based identification

```python
mztol=15
rttol=20
for idx1 in ft.index:
    
    feature_mzdiff=[]
    mz=ft.ix[idx1, "#feature ID"].split('_')[0]
    rt=ft.ix[idx1, "#feature ID"].split('_')[1]
    for idx2 in gnps.index:
        mzdiff= abs(float(mz)-float(gnps.ix[idx2, 'Precursor_MZ'])) 
        rtdiff= abs(float(mz)-float(gnps.ix[idx2, 'RT_Query']))
        if mzdiff <=mztol and rtdiff <=rttol:
            feature_mzdiff.append((gnps.ix[idx2, 'Compound_Name'], mzdiff))
    if feature_mzdiff==[]:
        continue
    else:
        sfeature_mzdiff=sorted(feature_mzdiff, key = lambda element : element[1])
        ft.ix[idx1, "#feature ID"]=sfeature_mzdiff[0][0]
```

## converting OTU IDs to taxonomy

```python
otu_tax={}
for otu in gg97.index:
    otu_tax[otu]=gg97.ix[otu,"taxonomy"]
```

```python
for otus in ft.columns[1:]:
    ft=ft.rename(columns={otus:otu_tax[int(otus)]+"id"+otus})
```
