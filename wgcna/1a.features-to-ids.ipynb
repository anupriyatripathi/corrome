## fuzzy conversion of MS1 features to MS2-based identification

```python
mztol=15
rttol=20
for ft in metabft.columns[1:]:
    
    feature_mzdiff=[]
    mz=ft.split('_')[0]
    rt=ft.split('_')[1]
    for idx in gnps.index:
        mzdiff= abs(float(mz)-float(gnps.ix[idx, 'Precursor_MZ'])) 
        rtdiff= abs(float(mz)-float(gnps.ix[idx, 'RT_Query']))
        if mzdiff <=mztol and rtdiff <=rttol:
            feature_mzdiff.append((gnps.ix[idx, 'Compound_Name'], mzdiff))
    if feature_mzdiff==[]:
        continue
    else:
        sfeature_mzdiff=sorted(feature_mzdiff, key = lambda element : element[1])
        metabft=metabft.rename(columns={ft:sfeature_mzdiff[0][0]+"_"+str(mz)[0:5]+"_"+str(rt)[0:5]})
```

```python
metabft.to_csv("./metabft_w_ids_srows.txt", sep='\t', index=False)
```

## converting OTU IDs to taxonomy

```python
otu_tax={}
for otu in gg97.index:
    otu_tax[otu]=gg97.ix[otu,"taxonomy"]
```

```python
for otus in microft.columns[1:]:
    microft=microft.rename(columns={otus:otu_tax[int(otus)]+"id"+otus})
```
